```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#load cleansed data
df_public_transport = pd.read_csv('/content/Cleansed_Data_v.1 (2).csv')
```

```python
#transform workplace closing information
df_public_transport.loc[df_public_transport['PublicTransportClosing'] == 0, 'PublicTransportClosing_Definition'] = 'No_Measure'
df_public_transport.loc[df_public_transport['PublicTransportClosing'] == 1, 'PublicTransportClosing_Definition'] = 'Recommend_Closing'
df_public_transport.loc[df_public_transport['PublicTransportClosing'] == 2, 'PublicTransportClosing_Definition'] = 'Require_Closing'
```

```python
#find daily case using shift function
df_public_transport['Lagged_Confirmed_Case']= df_public_transport.groupby(['Country','State'])['ConfirmedCases'].shift(1)
df_public_transport['New_Daily_Cases'] = df_public_transport['ConfirmedCases']- df_public_transport['Lagged_Confirmed_Case']
```

```python
df_public_transport['Date'] = pd.to_datetime(df_public_transport.Date, format='%Y-%m-%d')
df_public_transport['Month'] = df_public_transport['Date'].dt.strftime('%m')
df_public_transport['Week_Number'] = df_public_transport['Date'].dt.isocalendar().week
```

```python
#Select Specific Columns
df_public_transport = df_public_transport [['Country',
                                                   'State',
                                                   'Date',
                                                   'PublicTransportClosing',
                                                   'PublicTransportClosing_Definition',
                                                   'ConfirmedCases',
                                                   'New_Daily_Cases',
                                                   'Month',
                                                   'Week_Number'

                                                   ]]
```

```python
#Find max confirmed cases per state and save to CSV
df_Max_Cases_Per_State = df_public_transport.groupby('State')['ConfirmedCases'].max().reset_index().sort_values(by=['ConfirmedCases'],ascending = False)
df_Max_Cases_Per_State['Rank']= df_Max_Cases_Per_State['ConfirmedCases'].rank(ascending = False)
df_Max_Cases_Per_State.to_csv(r"Max_Confirmed_State_Per_Case.csv", index=False)
```
