```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#Load Cleansed Data
df_analysis_InternationalRestrictions = pd.read_csv('/content/Cleansed_Data.csv')
```

```python
#Transform InternationalRestrictions Info
df_analysis_InternationalRestrictions.loc[df_analysis_InternationalRestrictions['InternationalRestrictions'] == 0, 'InternationalRestrictionsDetail'] = 'No_Measure'
df_analysis_InternationalRestrictions.loc[df_analysis_InternationalRestrictions['InternationalRestrictions'] == 1, 'InternationalRestrictionsDetail'] = 'Screening_Arrivals'
df_analysis_InternationalRestrictions.loc[df_analysis_InternationalRestrictions['InternationalRestrictions'] == 2, 'InternationalRestrictionsDetail'] = 'Quarantine_Arrivals_Few'
df_analysis_InternationalRestrictions.loc[df_analysis_InternationalRestrictions['InternationalRestrictions'] == 3, 'InternationalRestrictionsDetail'] = 'Ban_Arrivals_Few'
df_analysis_InternationalRestrictions.loc[df_analysis_InternationalRestrictions['InternationalRestrictions'] == 4, 'InternationalRestrictionsDetail'] = 'Ban_Arrivals_All'
```

```python
df_analysis_InternationalRestrictions['Date'] = pd.to_datetime(df_analysis_InternationalRestrictions.Date, format= '%Y-%m-%d')
df_analysis_InternationalRestrictions['Month'] = df_analysis_InternationalRestrictions['Date'].dt.strftime("%m")
df_analysis_InternationalRestrictions['Week_Number'] = df_analysis_InternationalRestrictions['Date'].dt.isocalendar().week
```

```python
#Find Daily Case Using Shift()
df_analysis_InternationalRestrictions['Lagged_Confirmed_Case'] = df_analysis_InternationalRestrictions.groupby(['Country', 'State'])['ConfirmedCases'].shift(1)
df_analysis_InternationalRestrictions['Daily_New_Cases'] = df_analysis_InternationalRestrictions['ConfirmedCases'] - df_analysis_InternationalRestrictions['Lagged_Confirmed_Case']
```

```python
#Select Specific Columns
df_analysis_InternationalRestrictions = df_analysis_InternationalRestrictions[[
    'Country', 'State', 'Date', 'InternationalRestrictions', 'InternationalRestrictionsDetail', 'ConfirmedCases', 'Daily_New_Cases', 'Month', 'Week_Number'
]]
```

```python
#Find Max Confirmed Cases per State and Save to CSV
df_MaxCases_Per_State = df_analysis_InternationalRestrictions.groupby('State')['ConfirmedCases'].max().reset_index().rename(columns= {'Daily_New_Cases':'Weekly_Cases'}).sort_values(by=['ConfirmedCases'], ascending= False)
df_MaxCases_Per_State['Rank'] = df_MaxCases_Per_State['ConfirmedCases'].rank(ascending= False)
df_MaxCases_Per_State.to_csv(r"MaxConfirmedCasesPerState.csv", index= False)
```
