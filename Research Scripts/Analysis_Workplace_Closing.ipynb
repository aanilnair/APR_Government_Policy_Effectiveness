```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#load cleansed data
df_analysis_workplace = pd.read_csv('/content/Cleansed_Data_v.1 (2).csv')
```

```python
#transform workplace closing information
df_analysis_workplace.loc[df_analysis_workplace['WorkplaceClosing'] == 0, 'WorkplaceClosing_Definition'] = 'No_Measure'
df_analysis_workplace.loc[df_analysis_workplace['WorkplaceClosing'] == 1, 'WorkplaceClosing_Definition'] = 'Recommend_Closing'
df_analysis_workplace.loc[df_analysis_workplace['WorkplaceClosing'] == 2, 'WorkplaceClosing_Definition'] = 'Require_Closing_Few'
df_analysis_workplace.loc[df_analysis_workplace['WorkplaceClosing'] == 3, 'WorkplaceClosing_Definition'] = 'Require_Closing_All'
```

```python
#find daily case using shift function
df_analysis_workplace['Lagged_Confirmed_Case']= df_analysis_workplace.groupby(['Country','State'])['ConfirmedCases'].shift(1)
df_analysis_workplace['New_Daily_Cases'] = df_analysis_workplace['ConfirmedCases']- df_analysis_workplace['Lagged_Confirmed_Case']
```

```python
df_analysis_workplace['Date'] = pd.to_datetime(df_analysis_workplace.Date, format='%Y-%m-%d')
df_analysis_workplace['Month'] = df_analysis_workplace['Date'].dt.strftime('%m')
df_analysis_workplace['Week_Number'] = df_analysis_workplace['Date'].dt.isocalendar().week
```

```python
#Select Specific Columns
df_analysis_workplace = df_analysis_workplace [['Country',
                                                   'State',
                                                   'Date',
                                                   'WorkplaceClosing',
                                                   'WorkplaceClosing_Definition',
                                                   'ConfirmedCases',
                                                   'New_Daily_Cases',
                                                   'Month',
                                                   'Week_Number'

                                                   ]]
```

```python
#Find max confirmed cases per state and save to CSV
df_Max_Cases_Per_State = df_analysis_workplace.groupby('State')['ConfirmedCases'].max().reset_index().sort_values(by=['ConfirmedCases'],ascending = False)
df_Max_Cases_Per_State['Rank']= df_Max_Cases_Per_State['ConfirmedCases'].rank(ascending = False)
df_Max_Cases_Per_State.to_csv(r"Max_Confirmed_State_Per_Case.csv", index=False)
```
