```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#Load Cleansed Data
df_analysis_gathering = pd.read_csv('/content/Cleansed_Data.csv')
```

```python
#Changing Date Datatype
df_analysis_gathering = df_analysis_gathering.rename(columns = {'Date' : 'OldDate'})
df_analysis_gathering['Date'] = pd.to_datetime(df_analysis_gathering['OldDate'], format='%Y-%m-%d')
```

```python
#Transform Workplace Closing Info
df_analysis_gathering.loc[df_analysis_gathering['GatheringRestrictions'] == 0, 'GatheringDetail'] = 'No_Restriciton'
df_analysis_gathering.loc[df_analysis_gathering['GatheringRestrictions'] == 1, 'GatheringDetail'] = 'Restriction_1000+_people'
df_analysis_gathering.loc[df_analysis_gathering['GatheringRestrictions'] == 2, 'GatheringDetail'] = 'Restriction_1000_people'
df_analysis_gathering.loc[df_analysis_gathering['GatheringRestrictions'] == 3, 'GatheringDetail'] = 'Restriction_100_people'
df_analysis_gathering.loc[df_analysis_gathering['GatheringRestrictions'] == 4, 'GatheringDetail'] = 'Restriction_10_people'
```

```python
#Find Daily Case Using Shift()
df_analysis_gathering['Lagged_Confirmed_Case'] = df_analysis_gathering.groupby(['Country','State'])['ConfirmedCases'].shift(1)
df_analysis_gathering['Daily_New_Cases'] = df_analysis_gathering['ConfirmedCases'] - df_analysis_gathering['Lagged_Confirmed_Case']
```

```python
df_analysis_gathering['Date'] = pd.to_datetime(df_analysis_gathering.Date, format='%Y-%m-%d')
df_analysis_gathering['Month'] = df_analysis_gathering['Date'].dt.strftime('%m')
df_analysis_gathering['Week_Number'] = df_analysis_gathering['Date'].dt.isocalendar().week
```

```python
#Select Specefic Columns
df_analysis_gathering = df_analysis_gathering[[
    'Country',
    'State' ,
    'Date' ,
    'GatheringRestrictions',
    'GatheringDetail' ,
    'ConfirmedCases' ,
    'Daily_New_Cases' ,
    'Month',
    'Week_Number'
    ]]
```

```python
#Select Wisconsin
df_Wisconsin = df_analysis_gathering[((df_analysis_gathering['State']== 'Wisconsin'))]
df_Wisconsin = df_Wisconsin[[
    'Country',
    'State' ,
    'GatheringDetail' ,
    'Daily_New_Cases',
    'Month' ,
    'Week_Number'
    ]]
df_Wisconsin.groupby(['State', 'GatheringDetail', 'Month', 'Week_Number'])['Daily_New_Cases'].sum().reset_index().rename(columns= {'Daily_New_Cases':'Weekly_Cases'}).sort_values(by=['Month', 'Week_Number'],ascending= True)
```
