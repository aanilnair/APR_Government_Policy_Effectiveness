```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#Load Cleansed Data
df_analysis_school = pd.read_csv('/content/Cleansed_Data.csv')
```

```python
#Transform SchoolClosing Info
df_analysis_school.loc[df_analysis_school['SchoolClosing'] == 0, 'SchoolClosingDetail'] = 'No_Measure'
df_analysis_school.loc[df_analysis_school['SchoolClosing'] == 1, 'SchoolClosingDetail'] = 'Recommended_Closing'
df_analysis_school.loc[df_analysis_school['SchoolClosing'] == 2, 'SchoolClosingDetail'] = 'Required_Closing_Few'
df_analysis_school.loc[df_analysis_school['SchoolClosing'] == 3, 'SchoolClosingDetail'] = 'Required_Closing_All'
```

```python
df_analysis_school['Date'] = pd.to_datetime(df_analysis_school.Date, format= '%Y-%m-%d')
df_analysis_school['Month'] = df_analysis_school['Date'].dt.strftime("%m")
df_analysis_school['Week_Number'] = df_analysis_school['Date'].dt.isocalendar().week
```

```python
#Find Daily Case Using Shift()
df_analysis_school['Lagged_Confirmed_Case'] = df_analysis_school.groupby(['Country', 'State'])['ConfirmedCases'].shift(1)
df_analysis_school['Daily_New_Cases'] = df_analysis_school['ConfirmedCases'] - df_analysis_school['Lagged_Confirmed_Case']
```

```python
#Select Specific Columns
df_analysis_school = df_analysis_school[[
    'Country', 'State', 'Date', 'SchoolClosing', 'SchoolClosingDetail', 'ConfirmedCases', 'Daily_New_Cases', 'Month', 'Week_Number'
]]
```

```python
#Find Max Confirmed Cases per State and Save to CSV
df_MaxCases_Per_State = df_analysis_school.groupby('State')['ConfirmedCases'].max().reset_index().rename(columns= {'Daily_New_Cases':'Weekly_Cases'}).sort_values(by=['ConfirmedCases'], ascending= False)
df_MaxCases_Per_State['Rank'] = df_MaxCases_Per_State['ConfirmedCases'].rank(ascending= False)
df_MaxCases_Per_State.to_csv(r"MaxConfirmedCasesPerState.csv", index= False)
```
