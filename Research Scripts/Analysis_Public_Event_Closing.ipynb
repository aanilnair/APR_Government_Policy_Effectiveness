```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#load cleansed data
df_public_event = pd.read_csv('/content/Cleansed_Data_v.1 (2).csv')
```

```python
#transform workplace closing information
df_public_event.loc[df_public_event['PublicEventClosing'] == 0, 'PublicEventClosing_Definition'] = 'No_Measure'
df_public_event.loc[df_public_event['PublicEventClosing'] == 1, 'PublicEventClosing_Definition'] = 'Recommend_Cancelling'
df_public_event.loc[df_public_event['PublicEventClosing'] == 2, 'PublicEventClosing_Definition'] = 'Require_Cancelling'
```

```python
#find daily case using shift function
df_public_event['Lagged_Confirmed_Case']= df_public_event.groupby(['Country','State'])['ConfirmedCases'].shift(1)
df_public_event['New_Daily_Cases'] = df_public_event['ConfirmedCases']- df_public_event['Lagged_Confirmed_Case']
```

```python
df_public_event['Date'] = pd.to_datetime(df_public_event.Date, format='%Y-%m-%d')
df_public_event['Month'] = df_public_event['Date'].dt.strftime('%m')
df_public_event['Week_Number'] = df_public_event['Date'].dt.isocalendar().week
```

```python
#Select Specific Columns
df_public_event = df_public_event [['Country',
                                                   'State',
                                                   'Date',
                                                   'PublicEventClosing',
                                                   'PublicEventClosing_Definition',
                                                   'ConfirmedCases',
                                                   'New_Daily_Cases',
                                                   'Month',
                                                   'Week_Number'

                                                   ]]
```

```python
#Find max confirmed cases per state and save to CSV
df_Max_Cases_Per_State = df_public_event.groupby('State')['ConfirmedCases'].max().reset_index().sort_values(by=['ConfirmedCases'],ascending = False)
df_Max_Cases_Per_State['Rank']= df_Max_Cases_Per_State['ConfirmedCases'].rank(ascending = False)
df_Max_Cases_Per_State.to_csv(r"Max_Confirmed_State_Per_Case.csv", index=False)
```
